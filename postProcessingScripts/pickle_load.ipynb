```python
import code 
import pickle
import os
import numpy as np

path = "/home/tyty/Desktop/CFD_Files/AFRL_Scholars/1st_Round_Ls"
os.chdir(path)
dirList = list(os.listdir(path))
dirListRun = []
for entry in dirList: 
    if entry.startswith('Run'):
        dirListRun.append(entry)
dirList = None
dirListRun.sort()

#Initialize our new data structure, using the sizes of the pickled "fluidDict"
dirName = dirListRun[0]
path = "/home/tyty/Desktop/CFD_Files/AFRL_Scholars/1st_Round_Ls"
path += '/' + dirName
os.chdir(path)
F = open('./fluid_dict.pkl', 'rb')
fluidDict=pickle.load(F)

# This neat bit of code uses the locals()[] to turn a string into a variable so we can 
# iteratively create variables. 
varNames = [
    'x',
    'y',
    'z',
    'qw',
    'taux',
    'tauy',
    'tauz',
    'tauw',
    'yplus',
    'T',
    'p',
    'x_cc',
    'y_cc',
    'z_cc'
]
for name in varNames:
    locals()[name] = None
    locals()[name] = np.zeros((len(fluidDict[name]),len(dirListRun)))
    
connDict = {}

i = 0
for dirName in dirListRun: 
    fluidDict = None
    path = "/home/tyty/Desktop/CFD_Files/AFRL_Scholars/1st_Round_Ls"
    path += '/' + dirName
    os.chdir(path)
    F = open('./fluid_dict.pkl', 'rb')
    fluidDict=pickle.load(F)
    for name in varNames:
        locals()[name][:,i] = fluidDict[name]
    connDict[str(i+1)] = fluidDict["conn"] 
    i += 1

path = '/home/tyty/Desktop/CFD_Files/MFmodelTestData'
os.chdir(path)
for name in varNames:
    pickleName = './' + name + '.pkl'
    f = open(pickleName, 'wb')
    pickle.dump(locals()[name], f)
    f.close()

# conn is special since we can't store it in an array. 
pickleName = './' + 'conn' + '.pkl'
f = open(pickleName, 'wb')
pickle.dump(connDict, f)
f.close()
    
```
